```python
# Figure 객체 내에 그래프가 존재. 그래프를 위한 새로운 figure 생성
fig = plt.figure()
```

```python
# figure 크기 : 2*2, 4개의 서브플롯 중 첫 번째 선택
ax = fig.add_subplot(1, 1, 1)
ax
```

```python
ax.plot(data)
```

```python
ax.annotate('default line', xy=(5,5), xytext=(5,5),
            arrowprops={'arrowstyle': '-'}, va='center')
```
